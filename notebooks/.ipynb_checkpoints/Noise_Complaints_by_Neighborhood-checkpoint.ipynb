```python
import numpy as np
from operator import add
import time

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
import fiona
import fiona.crs
import shapely
import rtree
import pandas as pd
import geopandas as gpd
```

```python
## Function that spatially matches the neighborhood index and complaint counts using rtree

def Process_Complaints(partId, records):  # parts: list of records    #epsg=2263
    import csv 
    import time
    import rtree
    import fiona.crs
    import pyproj
    #import pyproj as proj
    import geopandas as gpd
    import shapely
    import shapely.geometry as geom
    
    if partId==0:  # skip the header--first line of fist partition
        records.next()
    
    counts = {}
    reader = csv.reader(records)

    proj = pyproj.Proj(init='epsg:2263', preserve_units=True)  #'epsg:{}'.format(2263)
    uarea_shpfile = "../../../Desktop/kzimanzhou_study/CUSP/Courses/17Spr-BigData/Project-NoiseComplaints/data/ZillowNeighborhoods-NY/ZillowNeighborhoods-NY.shp"
    shapefile = uarea_shpfile  # load aggregate unit area shapfile (i.e. neighbourhood)
    neighborhoods = gpd.read_file(shapefile).to_crs(fiona.crs.from_epsg(2263))
    index = rtree.Rtree()
    
    for idx, geometry in enumerate(neighborhoods.geometry):
        index.insert(idx, geometry.bounds)
    
    for row in reader:
        if (row[1] and row[5] and row[6] and row[51] and row[50]):
            (Id, create_date, c_type, desc, X_Lon, Y_Lat) = (row[0],row[1], row[5],row[6], float(row[51]), float(row[50]))
            print X_Lon, Y_Lat

            # struct time
            struc_datetime = time.strptime(create_date, '%m/%d/%Y %I:%M:%S %p')
            year = struc_datetime.tm_year
            weekday = struc_datetime.tm_wday
            hour = struc_datetime.tm_hour

            if year != 2016:
                continue

            ncLoc= geom.Point(proj(X_Lon, Y_Lat))  # Lon: float(row[51]),  Lat:float(row[50]
            potentialMatches = index.intersection((ncLoc.x, ncLoc.y, ncLoc.x, ncLoc.y))

            match = None
            for idx in potentialMatches:
                if neighborhoods.geometry[idx].contains(ncLoc):
                    match = idx
                    break
            if match:
                counts[match] = counts.get(match, 0)+ 1
    
    return counts.items()
```

```python
## aggregated area file
neibor_shpfile = "../../../Desktop/kzimanzhou_study/CUSP/Courses/17Spr-BigData/Project-NoiseComplaints/data/ZillowNeighborhoods-NY/ZillowNeighborhoods-NY.shp"
neighborhoods = gpd.read_file(neibor_shpfile).to_crs(fiona.crs.from_epsg(2263))

## item data
File_Noise_1016 = '../../../Desktop/kzimanzhou_study/CUSP/Courses/17Spr-BigData/Project-NoiseComplaints/data/311_Noise_Complaints_10-16.csv'
rdd_noise = sc.textFile(File_Noise_1016, use_unicode=True).cache()  # keep in memory for repeated readings
```

```python
### 2016
noise_neighbor = rdd_noise.mapPartitionsWithIndex(Process_Complaints).reduceByKey(lambda x,y: x+y)
#noise_neighbor.take(10)
counts = noise_neighbor.collect()

print counts
```
